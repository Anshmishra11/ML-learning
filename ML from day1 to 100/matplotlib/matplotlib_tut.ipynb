```python
# from a pandas dataframe 
batsman= pd.read_csv('sharma-kohli.csv') 
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#multiple as prallel
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#labels tittle
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('rohit sharma vs virat kohli carrer comparison')

plt.xlabel('season')
plt.ylabel('runs scored')
plt.legend(loc='best')
```

```python
#colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='brown')

plt.title('rohit sharma vs virat kohli carrer comparison')

plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',marker='>',markersize='8')
plt.plot(batsman['index'],batsman['RG Sharma'],color='brown',marker='D')

plt.title('rohit sharma vs virat kohli carrer comparison')

plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
#legend -> location

plt.plot(batsman['index'],batsman['V Kohli'],color='green',marker='>',markersize='8',label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='brown',marker='D',label='rohit')

plt.title('rohit sharma vs virat kohli carrer comparison')

plt.xlabel('season')
plt.ylabel('runs scored')


#grid
plt.grid()
```

## BAR CHART

```python
children=[10,20,40,10,5]
colors=['red','white','black','blue','pink'] 
plt.bar(colors,children)
```

```python
df=pd.read_csv('batsman_season_record.csv') 
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
```

# side by side multiple bar graph 

```python
plt.bar(np.arange(df.shape[0]) -0.2 ,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0])  ,df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2 ,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
 #stacked bar chart 
```

```python

plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2016']+df['2015'],label='2017')

plt.legend() 
plt.show()
```
