```python
import numpy as np 
import pandas as pd 
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python
df=pd.read_csv('covid_toy.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

```python
X_train
```

# 1.Aam zindagi

```python
#addinf simple imputer to fever col
si=SimpleImputer()
X_train_fever=si.fit_transform(X_train[['fever']])

# also the test data 

X_test_fever=si.fit_transform(X_test[['fever']])

X_train_fever.shape
```

```python
# ordinalenconding -> cough
oe=OrdinalEncoder(categories=[['Mild','Strong']])
X_train_cough=oe.fit_transform(X_train[['cough']]) 

#also the test data set
X_test_cough=oe.fit_transform(X_test[['cough']]) 
X_train_cough.shape
```

```python
# OneHotencoding -> Gender,city

ohe=OneHotEncoder(drop='first') 
X_train_gender_city=ohe.fit_transform(X_train[['gender','city']]) 

#also the test data

X_test_gender_city=ohe.fit_transform(X_test[['gender','city']]) 
X_train_gender_city.shape
```

```python
# extracting age
X_train_age = X_train.drop(columns=['gender', 'fever', 'cough', 'city']).values 
# Note: Changed from [['gender','fever','cough','city']] to ['gender','fever','cough','city']
# The double brackets were causing the error

# also the test data
X_test_age = X_test.drop(columns=['gender', 'fever', 'cough', 'city']).values

X_train_age.shape
```

# 2.COLUMN TRANSFORMER

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(), ['fever']),
    ('tnf2', OrdinalEncoder(categories=[['Mild', 'Strong']]), ['cough']),  # Fixed the parentheses here
    ('tnf3', OneHotEncoder(drop='first'), ['gender', 'city'])
], remainder='passthrough')
```

```python
transformer.fit_transform(X_train).shape
```
