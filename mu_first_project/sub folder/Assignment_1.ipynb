```python
import json 
```

```python
# LOAD THE DATA 
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)

    return data
```

```python
data=load_data("store_data.json")
print(data)

```

```python
 # clean and structure the data 

def clean_data(data):
    cleaned_data=[]
    unique_user=set()
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
    for user in data:
        raw_rating = str(user["rating"]).strip().lower()  # Added parentheses to strip method
        if raw_rating in text_to_num: 
            raw_rating = text_to_num[raw_rating]
        user["rating"]=raw_rating


        # handle  missind value

        raw_age =user.get("age")
        if(raw_age==None):
            user["age"]=None 

        #deduplication

        if(user["name"].strip() in unique_user):
            continue
        unique_user.add(user["name"])
        cleaned_data.append(user)
    

    return cleaned_data
         
```

```python
clean_data(data)
```

```python
# Get meaningful insights from data 
def get_insights(data):

    #avg rating
    tot_rating=0
    for user in data :
        tot_rating+= float(user["rating"])

    print(f"avg rating={tot_rating/len(data)}")


    # percentage of user with poor rating
    poor_rating=0
    for user in data:
        if(float(user["rating"])<3):
            poor_rating+=1
            

    print(f"% of user with poor rating ={poor_rating/len(data)}")
```

```python
get_insights(data)
```

```python
# build recommendation feature
def get_recommandation(data):
    recommendations=[]

    for user in data:
        cur_recomm={}
        cur_recomm["name"]=user["name"]

        if(float(user["rating"])>=4):
            cur_recomm["brand"]="Apple"
        else:
            cur_recomm["bramd"]="samsung"
    recommendations.append(cur_recomm)
    return recommendations
```

```python
get_recommandation(data)
```
